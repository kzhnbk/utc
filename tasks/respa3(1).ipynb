```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, roc_auc_score
from sklearn.ensemble import StackingClassifier
from sklearn.linear_model import LogisticRegression
import lightgbm as lgb
import xgboost as xgb
from catboost import CatBoostClassifier

# Load training data
train = pd.read_csv('/kaggle/input/problem-3-republic-olymp-tst-homework/train.csv')

df_pos = train[['description', 'conclusion']].copy()
df_pos['label'] = 1

df_neg = df_pos[['description', 'conclusion']].copy()
# shuffle conclusions to create negative pairs
while True:
    df_neg['conclusion'] = np.random.permutation(df_pos['conclusion'].values)
    if not (df_neg['conclusion'] == df_pos['conclusion']).any():
        break

df_neg['label'] = 0
# combine and shuffle
df_all = pd.concat([df_pos, df_neg], axis=0).sample(frac=1, random_state=42).reset_index(drop=True)
# combine text
df_all['text_pair'] = df_all['description'] + ' [SEP] ' + df_all['conclusion']

# train/validation split
X_train_text, X_val_text, y_train, y_val = train_test_split(
    df_all['text_pair'], df_all['label'],
    test_size=0.2, random_state=42, stratify=df_all['label']
)

# TF-IDF features
vectorizer = TfidfVectorizer(max_features=50000, ngram_range=(1,2))
X_train = vectorizer.fit_transform(X_train_text)
X_val = vectorizer.transform(X_val_text)

# Initialize individual models with GPU support
lgbm_clf = lgb.LGBMClassifier(
    objective='binary',
    boosting_type='gbdt',
    learning_rate=0.05,
    num_leaves=31,
    feature_fraction=0.8,
    bagging_fraction=0.8,
    bagging_freq=5,
    n_estimators=100,
    random_state=42,
    device='gpu'
)

xgb_clf = xgb.XGBClassifier(
    objective='binary:logistic',
    learning_rate=0.05,
    max_depth=-1,
    n_estimators=100,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42,
    use_label_encoder=False,
    eval_metric='logloss',
    tree_method = "hist",
    device = "cuda"
)

cat_clf = CatBoostClassifier(
    iterations=100,
    learning_rate=0.05,
    depth=-1,
    task_type='GPU',
    devices='0',
    verbose=False,
    random_seed=42
)

# Create stacking ensemble with logistic regression meta-model
stacking_clf = StackingClassifier(
    estimators=[
        ('lgbm', lgbm_clf),
        ('xgb', xgb_clf),
        ('cat', cat_clf)
    ],
    final_estimator=LogisticRegression(max_iter=1000),
    passthrough=False,
    cv=5,
    n_jobs=-1
)

# Train stacking ensemble
stacking_clf.fit(X_train, y_train)

# Validation performance
y_pred_proba = stacking_clf.predict_proba(X_val)[:, 1]
print("Validation ROC AUC:", roc_auc_score(y_val, y_pred_proba))
print("Validation Accuracy (threshold=0.5):", accuracy_score(y_val, (y_pred_proba > 0.5).astype(int)))

# Load test data
test_d = pd.read_csv('/kaggle/input/problem-3-republic-olymp-tst-homework/test_descriptions.csv')
test_c = pd.read_csv('/kaggle/input/problem-3-republic-olymp-tst-homework/test_conclusions.csv')
test_b = pd.read_csv('/kaggle/input/problem-3-republic-olymp-tst-homework/test_batches.csv')

# Build submission
submission = []
for _, batch in test_b.iterrows():
    desc_ids = eval(batch['description_ids'])
    conc_ids = eval(batch['conclusion_ids'])
    desc_texts = {d: test_d.loc[test_d['description_id'] == d, 'description'].values[0] for d in desc_ids}
    conc_texts = {c: test_c.loc[test_c['conclusion_id'] == c, 'conclusion'].values[0] for c in conc_ids}
    for d in desc_ids:
        pairs = [desc_texts[d] + ' [SEP] ' + conc_texts[c] for c in conc_ids]
        probs = stacking_clf.predict_proba(vectorizer.transform(pairs))[:, 1]
        best_idx = np.argmax(probs)
        submission.append({'description_id': d, 'conclusion_id': conc_ids[best_idx]})

# Save submission
df_sub = pd.DataFrame(submission)
df_sub.to_csv('submission.csv', index=False)
print("Saved submission.csv with GPU-enabled stacking ensemble")
```
