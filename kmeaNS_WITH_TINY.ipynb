```python
results = []
with torch.no_grad():
    for batch in tqdm(dataloader):
        inputs = batch[0].to('cuda')
        outputs = model(inputs)  # –ø—Ä–æ–≥–æ–Ω—è–µ–º —á–µ—Ä–µ–∑ –º–æ–¥–µ–ª—å
        results.append(outputs)

# –û–±—ä–µ–¥–∏–Ω—è–µ–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã
results = torch.cat(results, dim=0)
```

```python
from sklearn.cluster import KMeans
import numpy as np

kmeans = KMeans(n_clusters=32, init='k-means++', n_init=20, max_iter=300, random_state=42)

clusters = kmeans.fit_predict(data2kmeans_scaled)
```
